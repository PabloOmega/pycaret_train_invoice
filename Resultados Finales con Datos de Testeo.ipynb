```python
import numpy as np
import pandas as pd

data_test = pd.read_csv('Dataset/data_test.csv')
```

```python
data_test
```

```python
data_test = data_test.drop(["Unnamed: 0"], axis = 1)
```

```python
from pycaret.classification import *
```

```python
lgbm = load_model("tuned_lgbm2")
```

```python
lgbm
```

Se muestran los resultados. La columna prediction_label indica si es fraude o no, mientras que la columna prediction_score_1 indica la probabilidad de fraude.

```python
results = predict_model(lgbm, raw_score = True, data = data_test)
```

```python
results
```

```python
results.to_csv('Dataset/Resultados.csv')
```

Se ordenan los resultados por probabilidad

```python
results_sort = results.sort_values(by = "prediction_score_1", ascending = False)
results_sort
```

```python
results_sort.to_csv('Dataset/Resultados ordenados por Probabilidad.csv')
```
